```python
print(Fore.BLUE+"Welcome to Blue")     
print(Fore.BLUE+"Blue is a quiz that will help you learn some useful sentences for your daily life in Germany")
print(Fore.BLUE+"Type a or b and then the enter button")
print(Fore.BLUE+"Got it? Great let's start!")
class Question:
     def __init__(self, prompt, answer):
          self.prompt = prompt
          self.answer = answer

question_prompts = [
     "For Greeting we say: \n(a) Guten Tag\n(b) Wie geht's ",
     "Asking about a place: \n(a) Was kostet...?\n(b) Wo ist...?",
     "Thanking a person: \n(a) Bitte\n(b) Danke SchÃ¶n",
     "Asking about the price: \n(a) Was Kostet?\n(b) Wer ist das?",
     "Apologize: \n(a) Leider\n(b) Entschuldigung",
     
]

questions = [
     Question(question_prompts[0], "a"),
     Question(question_prompts[1], "b"),
     Question(question_prompts[2], "b"),
     Question(question_prompts[3], "a"),
     Question(question_prompts[4], "b"),
]

def run_quiz(questions):
    score = 0
    for question in questions:
          answer = input(question.prompt)
          if answer == question.answer:
               score +=1
               print(Fore.GREEN+"Correct! ")
          else:
              print(Fore.RED +"Oops Wrong Answer !")
    print(Fore.RED+"Game Over!")
    if (score>3):
         print(Fore.GREEN+"Good Job! you got", score, "out of", len(questions))
    elif (score==3):
         print(Fore.GREEN+"Nice Try! you got", score, "out of", len(questions))
    else:
         print(Fore.MAGENTA+"Sorry! you got", score, "out of", len(questions))
     
     
        
run_quiz(questions)
```
